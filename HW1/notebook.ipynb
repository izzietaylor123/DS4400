## Problem 4.2

```python
# create random matrices and compute their inverses
import numpy as np

a = np.random.randint(-10, 11, size=(3, 3))
b = np.random.randint(-10, 11, size=(3, 3))
c = np.random.randint(-10, 11, size=(3, 3))

print("a and a inverse:")
print(a, "\n\n", np.linalg.inv(a), "\n")
print("b and b inverse:")
print(b, "\n\n", np.linalg.inv(b), "\n")
print("c and c inverse:")
print(c, "\n\n", np.linalg.inv(c), "\n")
```

```python
print("Verification (matrix * inverse = identity):")
print("a * a_inv:")
print(np.matmul(a, np.linalg.inv(a)))
print("b * b_inv:")
print(np.matmul(b, np.linalg.inv(b)))
print("c * c_inv:")
print(np.matmul(c, np.linalg.inv(c)))
```

```python
# For each feature, write code to compute the average value, the min and max values, as well as its variance.

import pandas as pd

df = pd.read_csv('kc_house_data.csv')
df.drop(columns=['id', 'date', 'zipcode'], inplace=True)

price = df['price']
bedrooms = df['bedrooms']
bathrooms = df['bathrooms']
sqft_living = df['sqft_living']
sqft_lot = df['sqft_lot']
floors = df['floors']
waterfront = df['waterfront']
view = df['view']
condition = df['condition']
grade = df['grade']
sqft_above = df['sqft_above']
sqft_basement = df['sqft_basement']
yr_built = df['yr_built']
yr_renovated = df['yr_renovated']
lat = df['lat']
long = df['long']
sqft_living15 = df['sqft_living15']
sqft_lot15 = df['sqft_lot15']

features = [price, bedrooms, bathrooms, sqft_living, sqft_lot, floors, waterfront, view, condition, grade, sqft_above, sqft_basement, yr_built, yr_renovated, lat, long, sqft_living15, sqft_lot15]


for feature in features:
    print(f"Feature: {feature.name}")
    print(f"Average: {feature.mean()}")
    print(f"Min: {feature.min()}")
    print(f"Max: {feature.max()}")
    print(f"Variance: {feature.var()}\n")
```

```python
# Compute the correlation coefficient of each feature with the response. 
# Include a table with the correlation coefficient of each feature with the response. 

correlations = df.corrwith(df['price'])
print(correlations)
```
