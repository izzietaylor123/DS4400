```python
import numpy as np

a = np.array([[1, 7, 3], [9, 2, 5], [1, -4, 8]])
b = np.array([[2, 5, 9], [-1, -8, 0], [3, 7, 6]])
c = np.array([[9, 3, -4], [10, -7, 8], [1, 5, 2]])

print("a inverse:")
print(np.linalg.inv(a))
print("b inverse:")
print(np.linalg.inv(b))
print("c inverse:")
print(np.linalg.inv(c))

```

```python
print("\nVerification (matrix * inverse = identity):")
print("a * a_inv:")
print(np.matmul(a, np.linalg.inv(a)))
print("b * b_inv:")
print(np.matmul(b, np.linalg.inv(b)))
print("c * c_inv:")
print(np.matmul(c, np.linalg.inv(c)))
```
